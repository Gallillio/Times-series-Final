```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    non_zero_indices = y_true != 0
    return np.mean(np.abs((y_true[non_zero_indices] - y_pred[non_zero_indices]) / y_true[non_zero_indices])) * 100
```

```python
def evaluate_model(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mape = mean_absolute_percentage_error(y_true, y_pred)
    
    print("Mean Absolute Error (MAE): ", mae)
    print("Mean Squared Error (MSE): ", mse)
    print("Root Mean Squared Error (RMSE): ", rmse)
    print("Mean Absolute Percentage Error (MAPE): ", mape)
```

```python
df = pd.read_csv('jena_climate_2009_2016.csv')
```

```python
df['Date Time'] = pd.to_datetime(df['Date Time'], format='%d.%m.%Y %H:%M:%S')
```

```python
df.set_index('Date Time', inplace=True)
```

```python
print(df.head())
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['T (degC)'], label='Temperature (degC)')
plt.title('Temperature Time Series')
plt.xlabel('Date')
plt.ylabel('Temperature (degC)')
plt.legend()
plt.show()
```

```python
# Split the data into training and testing sets
train_size = int(len(df) * 0.8)
train, test = df['T (degC)'][:train_size], df['T (degC)'][train_size:]
```

```python
# Fit the ARIMA model on the training set
arima_model = ARIMA(train, order=(5, 0, 0))  # Example order (p=5, d=0, q=0) since data is stationary
arima_model_fit = arima_model.fit()
print(arima_model_fit.summary())
```

```python
start = len(train)
end = len(train) + len(test) - 1
arima_forecast = arima_model_fit.predict(start=start, end=end, typ='levels')

# Evaluate the ARIMA model
print("ARIMA Model Evaluation:")
evaluate_model(test, arima_forecast)
```

```python
# Fit the SARIMA model on the training set
sarima_model = ARIMA(train, order=(1, 0, 1), seasonal_order=(1, 1, 1, 12))
sarima_model_fit = sarima_model.fit()
print(sarima_model_fit.summary())
```

```python
sarima_forecast = sarima_model_fit.predict(start=start, end=end, typ='levels')
# Evaluate the SARIMA model
print("SARIMA Model Evaluation:")
evaluate_model(test, sarima_forecast)
```

```python
train_reshaped = np.arange(len(train)).reshape(-1, 1)
test_reshaped = np.arange(len(train), len(train) + len(test)).reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(train_reshaped, train)
```

```python
# Make predictions on the test set using Linear Regression
linear_forecast = linear_model.predict(test_reshaped)

# Evaluate the Linear Regression model
print("Linear Regression Model Evaluation:")
evaluate_model(test, linear_forecast)
```

```python
# Plot the original data, training set, test set, and forecasts from all models
plt.figure(figsize=(10, 6))
plt.plot(train, label='Training Set')
plt.plot(test, label='Test Set')
plt.plot(arima_forecast, label='ARIMA Forecast', color='red')
plt.plot(sarima_forecast, label='SARIMA Forecast', color='green')
plt.plot(test.index, linear_forecast, label='Linear Regression Forecast', color='blue')
plt.title('Temperature Time Series and Forecasts')
plt.xlabel('Date')
plt.ylabel('Temperature (degC)')
plt.legend()
plt.show()
```
