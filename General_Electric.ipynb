```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential, Model
from keras.layers import LSTM, Dense, GRU, Input, LayerNormalization, Dense, Conv1D, Lambda, Add, Activation
from sklearn.metrics import mean_absolute_error, mean_squared_error
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
# Load the time series data from the uploaded file
file_path = 'data/GE.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
print(f"Data row count: {data.shape[0]}")
```

```python
# Convert the 'Date' column to datetime and set it as index
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Plot the 'Close' price to visualize the time series
plt.figure(figsize=(14, 6))
plt.plot(data['Close'], label='Close Price')
plt.title('GE Stock Close Price History')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

```python
close_prices = data['Close'].values.reshape(-1, 1)

# Normalize the dataset
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(close_prices)
```

```python
# Split the data into training and testing sets
train_data_len = int(np.ceil(len(scaled_data) * 0.8))
train_data = scaled_data[:train_data_len]
test_data = scaled_data[train_data_len:]

# Create the training dataset
x_train = []
y_train = []
for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])
x_train = np.array(x_train)
y_train = np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))

# Create the testing dataset
x_test = []
y_test = close_prices[train_data_len:]
for i in range(60, len(test_data)):
    x_test.append(test_data[i-60:i, 0])
x_test = np.array(x_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
```

<h3> LSTM

```python
# LSTM model
LSTM_model = Sequential()
LSTM_model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1], 1)))
LSTM_model.add(LSTM(units=50, return_sequences=False))
LSTM_model.add(Dense(units=25))
LSTM_model.add(Dense(units=1))

# Compile
LSTM_model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
LSTM_model.fit(x_train, y_train, batch_size=1, epochs=10)

# Get the LSTM_model's predicted price values
predictions = LSTM_model.predict(x_test)
predictions = scaler.inverse_transform(predictions)

# Align the predictions with the actual test data
valid = data[train_data_len:]
valid = valid.iloc[60:]  # 60-day window

# Add the predictions to the 'valid' dataframe
valid['Predictions'] = predictions

print("done training")
```

<h3> TCN

```python
def tcn_block(model_layer, dilation_rate):
    prev_x = model_layer
    model_layer = Conv1D(filters=64, kernel_size=3, dilation_rate=dilation_rate, padding='causal', activation='relu')(model_layer)
    model_layer = Conv1D(filters=64, kernel_size=3, dilation_rate=dilation_rate, padding='causal')(model_layer)
    model_layer = Add()([model_layer, prev_x])
    model_layer = Activation('relu')(model_layer)
    return model_layer
```

```python
# Build the TCN model
inputs = Input(shape=(x_train.shape[1], 1))
model_layer = tcn_block(inputs, 1)
model_layer = tcn_block(model_layer, 2)
model_layer = tcn_block(model_layer, 4)
model_layer = tcn_block(model_layer, 8)
model_layer = tcn_block(model_layer, 16)
model_layer = tcn_block(model_layer, 32)
model_layer = tcn_block(model_layer, 64)
model_layer = Conv1D(filters=1, kernel_size=1)(model_layer)
outputs = Lambda(lambda tt: tt[:, -1, :])(model_layer)

model = Model(inputs=inputs, outputs=outputs)
model.compile(optimizer='adam', loss='mean_squared_error')

model.fit(x_train, y_train, batch_size=1, epochs=10)
# model.fit(x_train, y_train, batch_size=1, epochs=20, validation_split=0.2)

# Get the model's predicted price values
predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)

# Align the predictions with the actual test data
valid = data[train_data_len:]
valid = valid.iloc[60:]  # Align with the 60-day window

# Add the predictions to the 'valid' dataframe
valid['Predictions'] = predictions

print("done training")
```

```python
model = ARIMA(train_data, order=(5, 1, 0))  # order(p,d,q) where p=AR terms, d=differencing, q=MA terms
model_fit = model.fit()

# Make predictions
predictions = model_fit.forecast(steps=len(test_data_series))
predictions = pd.Series(predictions, index=test_data_series.index)

# Align the predictions with the actual test data
valid = pd.DataFrame({'Close': test_data_series, 'Predictions': predictions})
```

```python
# Plot the data
plt.figure(figsize=(14, 6))
plt.title('ARIMA Model')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.plot(train_data_series, label='Train')
plt.plot(valid['Close'], label='Val')
plt.plot(valid['Predictions'], label='Predictions', color='red')
plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
plt.show()

# Calculate the evaluation metrics
predictions = predictions.values  # Convert predictions to a NumPy array
actual = valid['Close'].values  # Convert actual values to a NumPy array

mae = mean_absolute_error(actual, predictions)
mse = mean_squared_error(actual, predictions)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((actual - predictions) / actual)) * 100

print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"MAPE: {mape}%")
```
